```python
# Encoding the values of species column 
encode = LabelEncoder()
iris.species = encode.fit_transform(iris.species)

# displaying the data set to check the encoded values 
iris
```

```python

plt.figure(figsize=(8,6))
sns.countplot(iris['species']);

print('The percentage of Iris-setosa class : %.2f' % (iris['species'].value_counts()[0] / len(iris)))
print('The percentage of Iris-virginica class : %.2f' % (iris['species'].value_counts()[1] / len(iris)))
print('The percentage of Iris-versicolor species class : %.2f' % (iris['species'].value_counts()[2] / len(iris)))
```

```python
# splitting the data set into train and test 

from sklearn.model_selection import train_test_split

train , test = train_test_split(iris,test_size=0.2,random_state=0)
```

```python
# describing the shape of data set

print('shape of training data : ',train.shape)

print('shape of testing data',test.shape)
```

```python
# seperate the target and independent variable

train_x = test.drop(columns=['species'],axis=1)

train_y = test['species']

# other methods for seperating the target and independent variable
#x_iris = iris['sepal_length', 'sepal_width', 'petal_lenth','petal_width']
#y_iris = iris['species']
#train_X= iris.iloc[:,:-1].values
#train_y= iris.iloc[:,4].values
```

```python
test_x = test.drop(columns=['species'],axis=1)

test_y = test['species']
```

```python
# displaying the train splitted data for predicting variables 
train_x
```

```python
# displaying the train splitted data for target variable species 
train_y
```

```python
# create the object of the model
model = LogisticRegression()

model.fit(train_x,train_y)

predict = model.predict(test_x)
```

```python
from sklearn.metrics import accuracy_score
print('Predicted Values on Test Data',encode.inverse_transform(predict))

print('Accuracy Score on test data :',accuracy_score(test_y,predict))
```
